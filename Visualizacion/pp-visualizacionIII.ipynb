```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
ataques_sns = pd.read_csv("../Pandas/ficheros_pandas/attacks3.csv", index_col=0)
ataques_sns.head(2)
```

En estos ejercicios aprenderemos a hacer algunas gráficas usando la librería seaborn. Igual que el ejercicio anterior de pair programming, para solucionar algunas preguntas tendréis que usar métodos y herramientas aprendidas anteriormente como groupby, value_counts, rename, filtrado de datos, etc.

1.- Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.


```python
ataques_2015 = ataques_sns[ataques_sns["year"] >= 2015]
ataques_2015.head(2)
```

```python
plt.figure(figsize=(12,10))

fig = sns.histplot(y= ataques_2015["activity"], data= ataques_2015, hue=ataques_2015["year"])



plt.show()
```

Podemos apreciar en esta gráfica que la actividad "surfing" sobresale de entre las demás en el año 2015.

2.- Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.

```python
plt.figure(figsize=(8,4))

fig = sns.histplot(x = ataques_2015["year"], data= ataques_2015, hue=ataques_sns["sex_"])

plt.show()
```

Apreciamos en el gráfico de barras que el género Masculino cuenta con una mayor frecuencia de ataques, que a medida que han pasado los años, ha ido bajando pero sigue siendo más frecuente que en los demás sexos. 
En el sexo femenino hubo un descenso en 2016 pero se volvió a incrementar en 2017.

3.- ¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.

```python
ataques_1500 = ataques_sns[ataques_sns["year"] >= 1500]
ataques_1500.head(2)
```

4.- Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
ataques_pais = ataques_sns[(ataques_sns["country"] == "USA") | (ataques_sns["country"] == "AUSTRALIA") | (ataques_sns["country"] == "SOUTH AFRICA")]
ataques_pais.head(2)
```

```python
ataques_1900 = ataques_pais[ataques_pais["year"] >= 1900]
ataques_1900.head(2)
```

```python
plt.figure(figsize=(15,7))

fig = sns.histplot(x = ataques_1900["year"] , data= ataques_1900, hue=ataques_1900["country"], bins=10)

plt.show()
```

Podemos apreciar que a partir de 1960 es la población de USA la que más ataques sufre. En los años anteriores era Australia quien registraba más ataques. También podemos apreciar que a medida que han pasado los años han incrementado los casos en los tres paises.

5.- Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
ataques_b = ataques_1900.groupby(["country", "year"])["case_number"].count().reset_index()
```

```python
plt.figure(figsize=(10,8))
fig = sns.boxplot(x="country", y="case_number" , data = ataques_b, palette="Blues")
plt.show()
```

En este caso vemos en la gráfica boxplot de los tres países (USA, AUSTRALIA, SOUTH AFRICA) el número de ataques en USA es mucho mayor que en los demás países. En este bloxplot nos realiza un conteo de los ataques producidos en cada uno de los países.
