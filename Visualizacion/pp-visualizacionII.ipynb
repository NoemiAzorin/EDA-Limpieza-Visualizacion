```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
visual_ataques = pd.read_csv("../Pandas/ficheros_pandas/attacks3.csv", index_col = 0)
visual_ataques.head(2)
```

1 - Ayer creamos un par de gráficas para USA y España, donde visualizamos el número de ataques en cada país a lo largo de los años y el tipo de los ataques. 

Hoy tendréis que hacer dos gráficas con subplots:

En la primera debéis mostrar los tipos ataques en USA y España, cada subplot corresponderá a un país. Debéis:
- Poner nombre a los ejes de cada uno de los subplots
- Poner un títulos a cada uno de los subplots
- Poner el color de las líneas en negro.
- Establecer el tamaño de la gráfica a 15 x 5
- Ponerle marcadores a las gráficas
- No incluir leyenda

```python
visual_spain = visual_ataques[visual_ataques["country"] == "SPAIN"]
visual_usa = visual_ataques[visual_ataques["country"] == "USA"]

visual_usa["type"].value_counts()
visual_tipo = pd.DataFrame(visual_usa["type"].value_counts())
visual_dict = dict(visual_tipo.iloc[:,0])

visual_spain["type"].value_counts()
visual_tipo_spain = pd.DataFrame(visual_spain["type"].value_counts())
visual_spain_dict = dict(visual_tipo_spain.iloc[:,0])
```

```python
fig,ax = plt.subplots(1,2 , figsize= (15,5))

ax[0].plot(visual_spain_dict.keys(), visual_spain_dict.values(), marker = "D", c = "black")
ax[1].plot(visual_dict.keys(), visual_dict.values(), marker = "D",  c = "black")

ax[0].set_xlabel("Tipos de ataques en España")
ax[0].set_ylabel("Frecuencia de ataques en España")

ax[1].set_xlabel("Tipos de ataques en USA")
ax[1].set_ylabel("Frecuencia de ataques en USA")

ax[0].set_title("Gráfico ataques España")
ax[1].set_title("Gráfico ataques USA")

plt.savefig("./Visualizacion-II/subplot_tipos.jpg")
plt.show()
```

En la segunda debéis mostrar los ataques por año para cada uno de los países.
- De nuevo tendréis que ponerle nombre a los ejes y título a cada gráfica.
- En este caso el subplot deberá tener dos filas y una columna
- No incluir leyenda
- Cambiar el color de las barras

```python
condicion2 = visual_spain[visual_spain["year"] >= 2000]
condicion1 = visual_usa[visual_usa["year"] >= 2000]


plt.figure()
plt.hist(condicion1["year"], facecolor="pink", edgecolor="green")
plt.xlabel("Años")
plt.ylabel("Numero de ataques en USA")


plt.figure()
plt.hist(condicion2["year"], facecolor="pink", edgecolor="green")
plt.xlabel("Años")
plt.ylabel("Numero de ataques en España")
```

```python
fig,ax = plt.subplots(2,1 , figsize= (17,10))

ax[0].hist(condicion1["year"], color = "blue", edgecolor = "red", linewidth = 3)
ax[1].hist(condicion2["year"], color = "red", edgecolor = "yellow", linewidth = 3)

ax[0].set_xlabel("Tipos de ataques por año en USA")
ax[0].set_ylabel("Frecuencia de ataques por año en USA")

ax[1].set_xlabel("Tipos de ataques por año en España")
ax[1].set_ylabel("Frecuencia de ataques por año en España")

ax[0].set_title("Gráfico ataques por año USA")
ax[1].set_title("Gráfico ataques por año en España")

plt.savefig("./Visualizacion-II/subplot_años.jpg")
plt.show()
```

2- Cread un único gráfico donde juntéis las gráficas de los tipos de ataques en una sola. Debéis:
- Poner leyenda a la gráfica
- Cada línea debe ir en un color
- Ponerle nombre a los ejes y a la gráfica
- Poner marcadores a las lineas

```python
plt.figure(figsize = (8, 4))

plt.plot(visual_spain_dict.keys(), visual_spain_dict.values(), marker = "*", c = "red")
plt.plot(visual_dict.keys(), visual_dict.values(), marker = "*",  c = "blue")

plt.xlabel("Tipos de ataques")
plt.ylabel("Frecuencia")

plt.legend(labels = ["España", "USA"])

plt.savefig("./Visualizacion-II/comparacion_tipos.jpg")
plt.show()
```
