Es el momento de poner en pr치ctica los groupby de pandas. Para esto vamos a trabajar con el csv creado en el ejercicio de pair programming anterior.

El objetivo de este ejercicio es agrupar los datos para contestar las siguientes preguntas:

```python
ataques_groupby = pd.read_csv("./ficheros_pandas/attacks3.csv", index_col = 0)
ataques_groupby.head(2)
```

1 - 쮺u치ntos ataques hubo por pa칤s? Devuelve un DataFrame.

游늷 NOTA Ordena los resultados de mayor a menor. Esto todav칤a no lo hemos visto, pero un DataFrame lo podemos ordenar con .sort_values(). [Aqu칤](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html) ten칠is una explicaci칩n de c칩mo hacerlo.

```python
country = ataques_groupby.groupby("country").value_counts()
df_ataques = pd.DataFrame(country)
df_ataques.reset_index(inplace=True)
df_ataques.sort_values(by="country", ascending = True)
df_ataques.head(5)
```

2 - 쮺u치ntos ataques ha habido por a침o? Devuelve un DataFrame. De nuevo ordena los resultados de mayor a menor.

```python
year = ataques_groupby.groupby("year").value_counts()
df_ataques = pd.DataFrame(year)
df_ataques.reset_index(inplace = True)
df_ataques.sort_values(by = "year", ascending = True)
df_ataques.head(5)
```

3 - 쮺u치ntos ataques hubo por a침o y sexo? Devolved un DataFrame solo del a침o 2000 en adelante.

```python
year_sex_2000 = df_ataques[df_ataques["year"] >= 2000]
year_sex_2000.head(2)
```

```python
year_sex = year_sex_2000.groupby(["year", "sex_"]).value_counts()
```

```python
year_sex_pd = pd.DataFrame(year_sex)
year_sex_pd.head(2)
```

```python
#eliminamos la columna que nos aparece como "0", para no tenerlas duplicadas y poder hacer un .reset_index()
year_sex_pd.drop(columns= 0, inplace = True)
```

```python
year_sex_pd.reset_index()
```

4 - Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. 쮺u치ntos de los ataques fueron letales por sexo? Devuelve un DataFrame.

```python
df_ataques_pais = df_ataques[(df_ataques["country"] == "USA") | (df_ataques["country"] == "AUSTRALIA") | (df_ataques["country"] == "SOUTH AFRICA") & (df_ataques["fatal_(y/n)"] == "Y")]
df_ataques_pais.head(2)
```

```python
df_letales = df_ataques_pais.groupby("sex_")["fatal_(y/n)"].count()
```

```python
df_letales_pais = pd.DataFrame(df_letales)
df_letales_pais.head()
```

```python
df_letales_pais.reset_index(inplace = True)
```

```python
df_letales_pais
```

5 - Usando el DataFrame creado en el ejercicio anterior. 쮼n qu칠 pa칤s hubo m치s ataques?

```python
df_final = df_ataques_pais.groupby("country").value_counts()
```

```python
df_final_pd =pd.DataFrame(df_final)
```

```python
#eliminamos la columna que nos aparece como "0", para no tenerlas duplicadas y poder hacer un .reset_index()
df_final_pd.drop(columns= 0, inplace = True)
```

```python
df_final_pd.reset_index(inplace=True)
df_final_pd.head(2)
```

```python
pais_mas_ataques = df_final_pd["country"].max()
pais_mas_ataques
```

```python
print(f"El pa칤s que ha recibido m치s ataques es: {pais_mas_ataques}")
```
