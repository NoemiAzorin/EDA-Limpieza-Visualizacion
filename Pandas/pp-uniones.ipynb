2.- Chequead que los nombres de las columnas ver si se llaman iguales entre dataframes . ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sean todas iguales unificad el nombre de las columnas.

```python
ataques_pair.columns
```

```python
ataques_historico.columns
```

```python
ataques_adicional.columns
```

```python
# Cambiaremos los dataframes historico y adicional para que estén todas las columnas en minúscula y sin espacios como en el primero.
```

```python
columnas_nuevas_historico = {col:col.lower().replace(" ", "_") for col in ataques_historico.columns}
```

```python
ataques_historico.rename(columns=columnas_nuevas_historico, inplace=True)
ataques_historico.head(2)
```

```python
columnas_nuevas_adicional = {col:col.lower().replace(" ", "_") for col in ataques_adicional.columns}
```

```python
ataques_adicional.rename(columns=columnas_nuevas_adicional, inplace=True)
ataques_adicional.head(2)
```

3.- En este momento nos habremos dado cuenta que el DataFrame que generamos en el pair programming Pandas IV es similar al DataFrame de los datos historicos. Si recordamos, eliminamos algunas columnas en la sesión III que no eran útiles, pero esas columnas están todavía en el histórico. Si no fuera por eso podríamos unir esos dataframes. Realizad los cambios que creais necesarios para que los dataframes sean iguales y luego juntadlos usando el mejor método.

```python
ataques_historico.drop(labels= ["unnamed:_0","unnamed:_22", "unnamed:_23", "case_number.1", "case_number.2","original_order", "investigator_or_source", "pdf", "href_formula"], axis='columns', inplace = True)
```

```python
ataques_pair.head(2)
```

```python
ataques_historico.head(2)
```

```python
ataques_historico.set_index(["case_number"], inplace=True)
ataques_historico.head(2)
```

```python
ataques_historico["siglo"] = np.where(ataques_historico["year"] < 2000, "siglo-XX", "siglo-XXI")
ataques_historico.head(2)
```

.- Unión dos dataframes : ataques_pair y ataques_historico

```python
ataques_pair.columns
```

```python
ataques_historico.columns
```

```python
ataques_concat = pd.concat([ataques_pair,ataques_historico], axis=0, join="outer")
```

```python
print(f"Comprobamos que se muestran los datos del csv histórico mostrando los años: {ataques_concat['year'].unique()}")
```

4.- Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el DataFrame de información adicional. Debatid que método es el más correcto para unir toda esta información y ejecutadlo.

📌 NOTA Cuando hagáis la unión considerad que la información que más nos interesa es la del DataFrame que creamos en el apartado anterior, lo cual determinará el tipo de unión.


```python
ataques_concat.head(2)
```

```python
ataques_adicional.head(2)
```

```python
ataques_merge = ataques_concat.merge(ataques_adicional, left_on="case_number", right_on="case_number")
```

```python
ataques_merge.head(2)
```
